Meta Data:
This dataset is laboratory record of Hepatitis C Virus test which contains 14 attributes.
1. Patient ID/No. 
2. Category (diagnosis) (values: '0=Blood Donor', '0s=suspect Blood Donor', '1=Hepatitis', '2=Fibrosis', '3=Cirrhosis')
3. Age
4. Sex
5. ALB - Albumin 
6. ALP - Alkaline Phosphatase 
7. ALT - Alanine Aminotransferase 
8. AST - Aspartate Aminotransferase 
9. BIL - Bilirubin
10. CHE - Acetylcholinesterase
11. CHOL - Cholesterol
12. CREA - Creatinine
13. GGT - Gamma-Glutamyl Transferase
14. PROT - Protein

Category  (blood donors vs. Hepatitis C (including its progress ('just' Hepatitis C, Fibrosis, Cirrhosis) is the Target feature

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.catplot(x="Category", data = df, kind = "count")
plt.savefig("img_31")
```

Inference:
- according to the above graph, we can say that maximum patients are suffering from Cirrhosis diseasein this dataset, followed by Hepatitis C and Fibrosis diseases.
- Least number of patients are there under Suspect Blood Donor Category

```python
sns.boxplot(x="Category", y="ALB", data = df)
```

Inference:
- A normal albumin range is 3.4 to 5.4 g/dL
- In this dataset, maximum blood donors are having Albumin level between 40-47 
- Suspect blood donors have ALB level less than 22
- Hepatitis C patients have ALB level between 43-48
- Fibrosis patients have ALB level between 39-47
- Cirrhosis patients have ALB level between 32-38

```python
sns.boxplot(x="Category", y="AST", data = df)
plt.savefig("img_32")
```

Inference:
- The reference range for aspartate aminotransferase (AST) is as follows: Males: 6-34 IU/L. Females: 8-40 IU/L.
- Blood donors have AST level between 20-40
- Suspect blood donors have AST level between 45-115
- Hepatitis C patients have AST level between 49-80
- Fibrosis patients have AST level between 50-110
- Cirrhosis patients have AST level between 70-120

```python
sns.catplot(x="Category", y="ALT", data = df)
```

Inference:
- The normal value for ALT in blood ranges from 29 to 33 units per liter (IU/L) for males and 19 to 25 IU/L for females, but this value can vary depending on the hospital. This range can be affected by certain factors, including gender and age.
- maximum blood donors have ALT level between 10-60
- suspect blood donors have ALT level between 10-20
- Hepatitis C patients have ALT level Between 10-30
- Fibrosis patients have ALT level between 10-35
- Cirrhosis patients have ALT level between 10-25

```python
sns.catplot(x="Category", y="ALP", data = df)
```

Inference:
- The normal range of ALP varies from person to person and depends on your age, blood type, gender, and whether you're pregnant. The normal range for serum ALP level is 20 to 140 IU/L , but this can vary from laboratory to laboratory
- Blood donors have ALP level between 40-100
- Suspect blood donors have ALP level between 80-125
- Hepatitis C patients have ALP level between 20-40
- Fibrosis patients have ALP level between 30-50
- Cirrhosis patients have ALP level between 30-110

```python
sns.catplot(x="Category", y="BIL", data = df)
plt.savefig("img_33")
```

Inference:
- It is normal to have some bilirubin in the blood. A normal level is: Direct (also called conjugated) bilirubin: less than 0.3 mg/dL (less than 5.1 mol/L) Total bilirubin: 0.1 to 1.2 mg/dL (1.71 to 20.5 mol/L)
- blood donors have BIL amount between 0-30
- suspect blood donors have BIL amount less than 15
- Hepatitis C patients have BIL amount between 10-30
- Fibrosis patients have BIL amount between 10-35
- Cirrhosis patients have BIL amount between 10-60 

```python
sns.catplot(x="Category", y="CHE", data = df)
plt.savefig("img_34")
```
